## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)


df_ConvertedComp['ConvertedComp'].plot(kind='hist', figsize=(8, 5))
plt.title('Histogram of Converted Compensation')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual USD')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Respondent, Age
FROM master
"""
df_Age = pd.read_sql_query(QUERY,conn)

df_Age['Age'].plot(kind='box', figsize=(10, 7))
plt.title("Box plots of Respondent's Age")
plt.ylabel('Interval Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
From master
"""
df_scatter=pd.read_sql_query(QUERY,conn)


df_scatter.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6))

plt.title('Age versus Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY="""
SELECT WorkweekHrs,CodeRevHrs,Age
from master
"""
df_bubble=pd.read_sql_query(QUERY,conn)

norm_Age = (df_bubble['Age'] - df_bubble['Age'].min()) / (df_bubble['Age'].max() - df_bubble['Age'].min())

df_bubble.plot(kind='scatter',
               x='WorkWeekHrs',
               y='CodeRevHrs',
               figsize=(14, 8),
               s=norm_Age * 1000 + 10)
plt.title("CodeRevHrs versus WorkWeekHrs of the Respondents")
plt.legend(['Age'], loc='upper right')
           
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY="""
SELECT *
from DatabaseDesireNextYear
"""
df_database=pd.read_sql_query(QUERY,conn)

df_groupby = df_database.groupby('DatabaseDesireNextYear', axis=0).count()
df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)

df_pie = df_groupby.head()
df_pie
```

```python
df_pie['Respondent'].plot(kind='pie',
                          figsize=(5, 6),
                          autopct='%1.1f%%')
plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY="""
SELECT WorkWeekHrs,CodeRevHrs,Age
from master
"""
df_stack=pd.read_sql_query(QUERY,conn)
df_stack = df_stack.loc[(df_stack['Age'] >= 30) & (df_stack['Age'] <= 35)].reset_index().drop(columns=['index'])
df_stack.head()
```

```python
import matplotlib.pyplot as plt

df_stack.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind='bar', stacked=True, figsize=(25, 10))
plt.xlabel('CodeRevHrs')
plt.ylabel('WorkWeekHrs')
plt.xticks(rotation=0)

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY="""
SELECT ConvertedComp, Age
from master
"""
df_line = pd.read_sql_query(QUERY,conn)
df_line = df_line.loc[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)].reset_index().drop(columns=['index'])
df_line.dropna(inplace=True)

df_line = df_line.groupby('Age', axis=0).median()

df_line.plot(kind='line')

plt.title('Average Compensation Based on Age')
plt.ylabel('Converted Compensation (Annual USD')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY="""
SELECT MainBranch, Respondent
from master
"""

df_bar=pd.read_sql_query(QUERY,conn)

df_bar = df_bar.groupby('MainBranch', axis=0).count()
df_bar.sort_values(["Respondent"], ascending=True, inplace=True)

df_bar.plot(kind='barh', figsize=(20, 5))
plt.xlabel('Total')
plt.title("MainBranch compared to Respondents")


plt.show()
```
